```python
# ---------------------------
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

# ---------------
fig, axes = plt.subplots(2, 3, tight_layout=True, figsize=(8, 6))

# ---------------------------
x = [0, 1, 2, 3]
y = [1, 2, -5, 2]

# [0, 0]
axes[0, 0].plot(x, y)
axes[0, 0].set_title('')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y', rotation=0, labelpad=20)  # Y

# ------------sin-----------------
x2 = np.linspace(0, 10, 100)
y2 = 2 + 2 * np.sin(2 * x2)

# [0, 1]sin
axes[0, 1].plot(x2, y2, linewidth=2.0)
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 11, 2),
               ylim=(-1, 5), yticks=np.arange(-1, 6))
axes[0, 1].set_title('sin')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y', rotation=0, labelpad=20)  # Y

# -------------------------------

x3 = np.random .normal(15, 5, 2000)

# [1,0]
axes[1, 0].hist(x3)
axes[1, 0].set_title('')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y', rotation=0, labelpad=20)  # Y

# -------------------------------

# 
np.random.seed(3)
x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))

# [1,1]
axes[1, 1].scatter(x4, y4)
axes[1, 1].set_title('')
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('Y', rotation=0, labelpad=20)  # Y

# -------------------------------
plt.show()
```
