```python
#ライブラリーをインポート  STEP1
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

#カテゴリーの準備
df = pd.read_csv('sample_pandas_6.csv')
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

#カテゴリー別の個数を表示  STEP2
df_category=df['カテゴリー'].value_counts()

# カテゴリーとカウントを分けてリストにする
categories = df_category.index.tolist()
counts = df_category.values.tolist()
data = {'カテゴリー': categories, '数': counts}
df_mix = pd.DataFrame(data)

#棒グラフの作成
ax = df_mix.plot(kind='bar', x='カテゴリー', y='数', title='カテゴリーの統計情報', legend=False, color='skyblue', figsize=(10, 6))

#ラベルの設定
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha='center')
ax.set_xlabel('商品名')
ax.set_ylabel('数', rotation=0, labelpad=20)

#グラフの表示 STEP3
plt.show()

#商品番号ごとにグループ化し、注文数の基本統計量を計算する  STEP4
order_stats = df.groupby('商品番号')['注文数'].describe()

# 結果を表示する
print(order_stats)
```
